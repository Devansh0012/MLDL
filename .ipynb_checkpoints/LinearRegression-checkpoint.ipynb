```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
height = np.array([5.4,5.2,5.6,6.1,6.3])
weight = np.array([125,130,135,140,150])
```

```python
plt.scatter(height,weight)
plt.xlabel('Height(inches)')
plt.ylabel('Weight(pounds)')
plt.show()
```

```python
m, b = np.polyfit(height, weight, 1)
```

```python
plt.scatter(height,weight)
plt.scatter(height, m*height + b, color = 'red')
plt.xlabel('Height(inches)')
plt.ylabel('Weight(pounds)')
plt.show()
```

```python
new_height = 5.8
predicted_weight = m*new_height + b
print('predicted weight: ',predicted_weight)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('car_purchasing_data.csv')
data
```

```python
X = data[['gender','age','salary','debt','worth']]
y = data[['amount']]
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
new_data = np.array([[1,33,75835,15200,385190]])
predicted_price = model.predict(new_data)
print('purchasing price will be: ',predicted_price)
```
